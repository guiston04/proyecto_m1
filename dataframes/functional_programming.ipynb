```python
# Pipeline funcional

def funcion_guarra():
    df = pd.read_csv('./Top5000.csv')   # Funcion para importar datos
    df = df[['ars_name', 'album', 'avg_rat', 'rel_date', 'gens', 'duration_ms']] # Funcion para seleccionar columnas
    df = df.sort_values(by='avg_rat', ascending=False) # Funcion para ordenar datos
    df['duration_ms'] = pd.to_timedelta(df['duration_ms'], unit='ms') # Funcion limpiar datos
    df['gens'] = df['gens'].apply(main_genre) # Funcion para limpiar datos
    df = df[df['gens'] == 'Hard Rock'] # Funcion para seleccionar datos
    df = df.reset_index().drop(['index'], axis = 1)
    return df   
```

```python
#Modulo acq

def acquisition(path):
    try:
        df = pd.read_csv(path)
        return df
    except:
        'No data'
```

```python
# Modulo Pepe

def col_select(df, cols):
    df = df[cols]
    return df
```

```python
def sort_df(df, col, order):
    # col = string, order = True/False
    df = df.sort_values(by=col, ascending=order).reset_index(drop=True)
    return df
```

```python
def delta_time(df, col_input, col_output='delta_time', unit='ms'):
    df[col_output] = pd.to_timedelta(df[col_input], unit=unit)
    return df
```

```python
def main_genre(genres):
    try:
        return genres.split(',')[0].strip()
    except:
        return 'no_genre'
```

```python
def genre_selector(df, col_input='gens', col_output='single_genre'):
    df[col_output] = df[col_input].apply(main_genre)
    return df
```

```python
def genre_filter(df, col, genre):
    if genre in list(df[col].unique()):
        df = df[df[col] == genre].reset_index(drop=True)
        return df
    else:
        return 'No correct genre selected'
```

```python
path1 = './Top5000.csv'
cols1 = ['ars_name', 'album', 'avg_rat', 'rel_date', 'gens', 'duration_ms']
col1 = 'avg_rat'
order1 = True
col_input1 = 'duration_ms'
col2 = 'gens'
genre1 = 'Hard Rock'
col3 = 'duration_ms'
order2 = True

df1 = acquisition(path1)
df2 = col_select(df1, cols1)
df3 = sort_df(df2, col1, order1)
df4 = delta_time(df3, col_input1)
df5 = genre_selector(df4, col_output='gens')
df6 = genre_filter(df5, col2, genre1)
df7 = sort_df(df6, col3, order2)
df7
```
